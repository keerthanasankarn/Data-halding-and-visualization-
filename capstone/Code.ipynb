```python
#To visualise Pair plots of Health Metrices
import seaborn as sns# Create a custom color palette
import matplotlib.pyplot as plt
import pandas as pd
data=pd.read_csv('data.csv')
# Create a pair plot with selected columns
selected_columns = ["Age", "Sleep Duration", "Physical Activity Level", "Stress Level", "Heart Rate"]
sns.pairplot(data[selected_columns])
plt.suptitle("Pair Plot of Health Metrics", y=1.02)
plt.show()
```

```python
#To measure sleep disorder

data["Sleep Disorder"].value_counts()
sns.countplot(x='Sleep Disorder', data = data)
plt.show()
```

```python
sns.countplot(x='Gender', data = data, hue='Sleep Disorder')
plt.show()
```

```python
sns.countplot(x='BMI Category', data=data, hue='Sleep Disorder' )
```

```python
stress_by_occupation = data.groupby('Occupation', as_index=False)['Stress Level'].mean()
stress_by_occupation
plt.figure(figsize=(12, 6))
sns.barplot(x='Stress Level', y='Occupation', data=stress_by_occupation.sort_values('Stress Level', ascending=False))
plt.title('Average Stress Level by Occupation')
plt.xlabel('Average Stress Level')
plt.ylabel('Occupation')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.regplot(x='Stress Level', y='Quality of Sleep', data=data, scatter_kws={'color':'blue'},line_kws={'color':'red'})
plt.title('Correlation between Stress Level and Quality of Sleep')
plt.xlabel('Stress Level')
plt.ylabel('Quality of Sleep')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.regplot(x='Quality of Sleep', y='Sleep Duration', data=data, scatter_kws={'color':'blue'},line_kws={'color':'red'})
plt.title('Correlation between Sleep Duration and Quality of Sleep')
plt.xlabel('Quality of Sleep')
plt.ylabel('Sleep Duration')
plt.show()
```

```python
sleep_list = ['None', 'Sleep Apnea', 'Insomnia']

fig, axes = plt.subplots(1, len(sleep_list), figsize=(16, 6), sharey=True)

for i, sleep_disorder in enumerate(sleep_list):

    ax = axes[i]

    sns.histplot(data=data[data['Sleep Disorder'] == sleep_disorder],
                 x='Age',
                 kde=True,
                 ax=ax)

    ax.set_title(sleep_disorder)


plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x="Physical Activity Level", y="Stress Level", data=data,errorbar=None )
plt.title("Stress Level by Physical Activity Level")
plt.xlabel("Physical Activity Level")
plt.ylabel("Stress Level")
plt.xticks(rotation=45)
plt.show()
```

```python
data2= data[['Sleep Duration','Quality of Sleep', 'Physical Activity Level','Stress Level','Heart Rate','Daily Steps']]

data2.corr()

correlation_matrix= data2.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True)
```
